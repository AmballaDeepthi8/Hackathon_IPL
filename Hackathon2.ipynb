```python
#13.  What is the strike rate of Kohli in 2016

#HINT: strike rate = (Total Runs / Total Balls Faced) * 100
```

```python
df4["batsman"].value_counts()
```

```python
df_k=df4[(df4["season"]==2016) & (df4["batsman"]=='V Kohli')]
```

```python
df_k.columns
```

```python
total_runs = df_k['total_runs'].sum()
total_balls = df_k['ball'].sum()
```

```python
strike_rate =(total_runs / total_balls) * 100
```

```python
strike_rate
```
